```python
import os, xml.etree.ElementTree as ET
import pandas as pd
```

```python
my_path = os.path.join('.', 'ANNOTATIONS')

records = []
for fname in os.listdir(my_path):
    if not fname.endswith('.xml'):
        continue
    tree = ET.parse(os.path.join(my_path, fname))
    root = tree.getroot()
    for obj in root.findall('.//object'):
        label = obj.find('name').text
        records.append({
            'annotation_path': os.path.join(my_path, fname),
            'image_path': os.path.join('.', 'IMAGES', fname.replace('.xml', '.jpg')),
            'label': label
        })
data = pd.DataFrame(records)
```
